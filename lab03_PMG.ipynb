# Question 1

Find a DEM dataset in the Earth Engine Data Catalog, clip the DEM to your home state, and display the DEM with a proper color palette on the map. For example, the sample map below shows the DEM of the state of Colorado.

```python
Map1 = geemap.Map()

# Load borders
borders = ee.FeatureCollection("FAO/GAUL/2015/level1")
region = borders.filter(ee.Filter.eq('ADM1_NAME', 'Castilla y LeÃ³n'))

# Load DEM
dataset = ee.Image("NASA/NASADEM_HGT/001")
elevation = dataset.select('elevation')

# Clip it
dem_region = elevation.clip(region)

# Config. visualization
vis_params = {
    'min': 0,
    'max': 2500,
    'palette': ['006633', 'E5FFCC', '662A00', 'D8D8D8', 'F5F5F5']
    }

# Add text to the map
text = "Created by Pablo Miranda"
Map1.add_text(text, fontsize=20, position='bottomright')

# set style border
style = {'color': 'ff0000ff', 'width': 2, 'lineType': 'solid', 'fillColor': 'FF000000'}

# Add Layers
Map1.centerObject(region, 7)
Map1.addLayer(region.style(**style), {}, 'cyl')
Map1.addLayer(dem_region, vis_params, 'elevation')
Map1.add_colorbar(vis_params, label="Elevation (m)", layer_name='SRTM DEM',
                 orientation='vertical', transparent_bg=False, font_size=15)

# Display the map
Map1

```

# Question 2

Use Sentinel-2 or Landsat-9 data, create a cloud-free imagery of the year 2021 for your home state. Display the imagery on the map with a proper band combination. For example, the sample map below shows a cloud-free false-color composite of Sentinel-2 imagery of the year 2021 for the state of Colorado.

```python
Map2 = geemap.Map()

# Load borders
borders = ee.FeatureCollection("FAO/GAUL/2015/level1")
region = borders.filter(ee.Filter.eq('ADM1_NAME', 'Castilla y LeÃ³n'))

# Load Sentinel 2 data
collection = (
    ee.ImageCollection('COPERNICUS/S2_SR')
    .filterDate('2021-01-01', '2022-01-01')
    .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 5))
)

# Create cloud-free imagery
image = collection.median()

# Clip it
cloudfree = image.clip(region)

vis = {
    'min': 0.0,
    'max': 3000,
    'bands': ['B8', 'B4', 'B3'],
}

# set style border
style = {'color': '000000ff', 'width': 2, 'lineType': 'solid', 'fillColor': 'FF000000'}

# Add text to the map
text = "Created by Pablo Miranda"

# Add Layers
Map2.add_text(text, fontsize=20, position='bottomright')
Map2.addLayer(region.style(**style), {}, 'cyl')
Map2.centerObject(region, 7)
Map2.addLayer(cloudfree, vis, 'Sentinel-2')
Map2
```
